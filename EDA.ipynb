```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Rating'].dropna(), kde=True, bins=20, color='red')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.show()
```

```python
Q1=data['Rating'].quantile(0.25)
Q3=data['Rating'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q1+1.5*IQR
filterdata =data[(data['Rating']>=lower_bound)&(data['Rating']<=upper_bound)]
plt.figure(figsize=(10, 6))
sns.boxplot(x='Rating',data=filterdata,color='green')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Category', data=data, hue='Category', order=data['Category'].value_counts().index, palette='viridis', legend=False)
plt.title('Count of Apps by Category')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```
